# Wrangling steps

- removing only one row duplicated in ID column
- removing irrelevent columns like(IMDB_UD - Cast - Homepage - tagline - overview - cast - production_companies - revenue - budget)
- Replace null values in (genres) column to (not mentioned)
- change type of (release_date) columns to date_time

```python
import numpy as np
import pandas as pd

df = pd.read_csv('movies.csv')

df.head(3)
```

```python

df.info()
```

```python
df.isnull().sum()

# number of null values in each column
```

```python
# delete irrelevent columns

df.drop(['imdb_id' , 'cast', 'homepage','tagline', 'overview' ,'original_title'
   ,'production_companies', 'revenue','keywords','budget','director'],axis = 1 , inplace = True)
```

```python
df.info()
```

```python
# remove only one non-unique id

df['id'].duplicated().sum() # = 1
df.drop_duplicates(inplace = True)
df.info()
```

```python
# rename revenue_adj to revenue, and budget_adj to budget

df.rename(columns = {'revenue_adj': 'Revenue'}, inplace = True)
df.rename(columns = {'budget_adj': 'Budget'} , inplace = True)
```

```python
# replace null names of directors to (unknown)
# replace null values in genres to (none)

df['genres'].fillna('NoNe' , axis = 0 , inplace = True)
```

```python
df.info()
```

```python
# convert release date to date_time

df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
# now the data is cleaned and ready for analysis

print(df.info())
df.head(3)
```

```python
import matplotlib.pyplot as plt

df.hist(figsize = (40,40))
```

# what's the relation between revenue and vote counts ?
